### 小試身手
計算 1 到 100 之間的和，如果值超過500就停止執行

```python
sum = 0
THRESHOLD = 5500

for i in range(1,101):
    sum += i
    if sum >= THRESHOLD:
        print(f"final number:{sum} stop at {i}")
        break
    else:
        print(f"sum:{sum}") 
```

計算 1 到 100 之間的奇數之和，但要跳過所有被 5 整除的數

```python
#ctrl shift p -> reload window
odd_sum = 0    
for i in range(1,100,2):
    if (i % 5) == 0:
        print(f"odd_sum:{odd_sum} skip {i}")
        continue
    odd_sum += i   
    print(f"odd sum:{odd_sum}")
```
