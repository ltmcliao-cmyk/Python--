### `json` 套件

| 用法 | 說明 |
|:--:|:--:|
|json.dump(<obj>,<file>) |物件轉JSON並寫入文件|
|json.dumps(<obj>) |物件轉JSON字串|
| json.load(f)|讀取JSON文件|
|json.loads(f) |將JSON字串轉成Python物件|

```python
#list、dict是不同程式語言資料傳遞的格式
#tuple、set不能作為json格式傳遞，其他程式語言未必有
import json
nums = [i for i in range(1,51) if i % 2 == 0]
print(nums)

with open('result.json',"w") as f:
    f.write(json.dumps(nums))
    
```
