百貨公司周年慶 消費金額超過 2000 要折 200  

請撰寫一支程式  
計算客戶消費的結帳金額

```python
THRESHOLD = 2000
DISCOUNT = 200

def discount_calculator(total):
    multiplier = total // THRESHOLD
    if multiplier >= 1:
        total -= multiplier * DISCOUNT
    return(total)
total = int(input("amount:"))
discount_calculator(total)
    
```
